### 10. Tạo ngẫu nhiên hai tensor 1-chiều x, y có 5 phần tử. Trả về x + y nếu x < y, và x – y nếu x > y

```python
x = tf.random.uniform(shape=[5], minval=0, maxval=10)
y = tf.random.uniform(shape=[5], minval=0, maxval=10)

# Sử dụng tf.where để thực hiện logic điều kiện theo từng phần tử
result_10 = tf.where(x < y, x + y, x - y)

print("x:", x.numpy())
print("y:", y.numpy())
print("Kết quả:", result_10.numpy())
```

### 11. Tạo ngẫu nhiên 2 tensor x = (3, 2) [5, 10], y = (2, 3) [15, 20]. Giữ giá trị đường chéo của x và thực hiện x + y.

```python
# Lưu ý: x + y chỉ thực hiện được khi hai ma trận có cùng kích thước hoặc có thể broadcast.
# Đề bài yêu cầu x(3,2) và y(2,3), có lẽ ý bạn là nhân ma trận hoặc cộng sau khi chuyển vị.
x_11 = tf.random.uniform(shape=[3, 2], minval=5, maxval=10)
y_11 = tf.random.uniform(shape=[2, 3], minval=15, maxval=20)

# Giả sử thực hiện x + y_transpose để cùng kích thước (3, 2)
y_t = tf.transpose(y_11)
result_add = x_11 + y_t

print("x_11:\n", x_11.numpy())
print("y_transpose:\n", y_t.numpy())
print("Kết quả cộng:\n", result_add.numpy())
```

### 12. Viết hàm tạo 2 tensor ngẫu nhiên kích thước (3, 2, 2) và các hàm cộng, trừ, nhân, chia trong graph

```python
@tf.function
def tensor_operations(x, y):
    add_res = tf.add(x, y)
    sub_res = tf.subtract(x, y)
    mul_res = tf.multiply(x, y)
    div_res = tf.divide(x, y)
    return add_res, sub_res, mul_res, div_res

x_12 = tf.random.uniform(shape=[3, 2, 2])
y_12 = tf.random.uniform(shape=[3, 2, 2])

# Thực thi hàm trong graph
a, s, m, d = tensor_operations(x_12, y_12)
print("Kết quả cộng (3, 2, 2):\n", a.numpy())
```

### 13. Xây dựng model để tính tổng hai tensor x, y

```python
class AddModel(tf.Module):
    @tf.function
    def __call__(self, x, y):
        return x + y

model_add = AddModel()
result_13 = model_add(x_12, y_12)
print("Tổng x + y qua model:", result_13.shape)
```

### 14. Xây dựng model để tính hiệu và tích hai tensor x, y

```python
class MathModel(tf.Module):
    @tf.function
    def compute(self, x, y):
        diff = x - y
        prod = x * y
        return diff, prod

model_math = MathModel()
diff, prod = model_math.compute(x_12, y_12)
print("Hiệu:", diff.shape, "Tích:", prod.shape)
```

### 15. Xây dựng model để tính chu vi hình chữ nhật P = (a + b) * 2

```python
class PerimeterModel(tf.Module):
    @tf.function
    def calculate(self, a, b):
        return (a + b) * 2

# Giả sử a, b là chiều dài và rộng
a = tf.constant([10.0, 20.0])
b = tf.constant([5.0, 15.0])
model_p = PerimeterModel()
p = model_p.calculate(a, b)
print("Chu vi hình chữ nhật:", p.numpy())
```

1.Tạo mảng 20 số nguyên ngẫu nhiên kích thước 3 x 3 có giá trị trong khoảng [0, 20]. Viết
chương trình kiểm tra tất cả các phần tử trong mảng khác 0 (tìm hiểu hàm all của numpy)

```python
import numpy as np

matrix = np.random.randint(0, 21, size=(3, 3))

print("Mảng ngẫu nhiên:")
print(matrix)

# Kiểm tra tất cả phần tử khác 0
# matrix != 0 trả về mảng Boolean, np.all kiểm tra nếu tất cả là True
check_nonzero = np.all(matrix != 0)

if check_nonzero:
    print("\nKết quả: Tất cả các phần tử trong mảng đều khác 0.")
else:
    print("\nKết quả: Có ít nhất một phần tử bằng 0 trong mảng.")
```

2.Tạo mảng 20 số nguyên ngẫu nhiên kích thước 3 x 3 có giá trị trong khoảng [0, 20]. Viết
chương trình kiểm tra tồn tại một phần tử trong mảng khác 0 (tìm hiểu hàm any của numpy)

```python
import numpy as np

# Tạo mảng 3x3 ngẫu nhiên [0, 20]
matrix = np.random.randint(0, 21, size=(3, 3))

print("Mảng ngẫu nhiên:")
print(matrix)

# Kiểm tra tồn tại ít nhất một phần tử khác 0
# matrix != 0 trả về mảng Boolean, np.any kiểm tra nếu có ít nhất một True
has_nonzero = np.any(matrix != 0)

if has_nonzero:
    print("\nKết quả: Có ít nhất một phần tử trong mảng khác 0.")
else:
    print("\nKết quả: Tất cả các phần tử đều bằng 0.")
```

3.Sinh viên tìm hiểu và viết chương trình sử dụng các hàm sau numpy.greater,
numpy.greater_equal, numpy.less, và numpy.less_equal để so sánh 2 mảng 1 chiều, 2 chiều.

```python
import numpy as np

# 1. So sánh mảng 1 chiều
a1 = np.array([10, 20, 30, 40])
b1 = np.array([15, 20, 25, 45])

print("--- So sánh mảng 1 chiều ---")
print(f"a1: {a1}")
print(f"b1: {b1}")
print("greater (a1 > b1):      ", np.greater(a1, b1))
print("greater_equal (a1 >= b1):", np.greater_equal(a1, b1))
print("less (a1 < b1):         ", np.less(a1, b1))
print("less_equal (a1 <= b1):   ", np.less_equal(a1, b1))

# 2. So sánh mảng 2 chiều
a2 = np.array([[1, 5], [8, 12]])
b2 = np.array([[2, 5], [7, 10]])

print("\n--- So sánh mảng 2 chiều ---")
print("Mảng a2:\n", a2)
print("Mảng b2:\n", b2)
print("greater (a2 > b2):\n", np.greater(a2, b2))
print("greater_equal (a2 >= b2):\n", np.greater_equal(a2, b2))
print("less (a2 < b2):\n", np.less(a2, b2))
print("less_equal (a2 <= b2):\n", np.less_equal(a2, b2))
```

4.Viết chương trình tạo mảng gồm 10 số 1, 10 số 0 và 10 số 5

```python
import numpy as np

# Tạo mảng gồm 10 số 0
zeros = np.zeros(10)

# Tạo mảng gồm 10 số 1
ones = np.ones(10)

# Tạo mảng gồm 10 số 5
fives = np.full(10, 5)

# Kết hợp các mảng lại
result = np.concatenate([ones, zeros, fives])

print("Mảng kết quả:")
print(result)
print("Kích thước mảng:", result.size)
```

### 8. Tạo mảng 20 phần tử [0, 20] và đổi dấu các số trong khoảng [9, 15]

```python
arr_20 = np.random.randint(0, 21, 20)
print("Mảng ban đầu:", arr_20)

# Đổi dấu các phần tử có giá trị từ 9 đến 15
arr_20[(arr_20 >= 9) & (arr_20 <= 15)] *= -1

print("Mảng sau khi đổi dấu:", arr_20)
```

### 10. Tạo ma trận 10 x 10 với biên là 1 và bên trong là 0

```python
# Tạo ma trận toàn số 1
border_matrix = np.ones((10, 10))

# Gán các phần tử bên trong (từ chỉ số 1 đến -1) bằng 0
border_matrix[1:-1, 1:-1] = 0

print("Ma trận 10x10 với biên bằng 1:")
print(border_matrix)
```

### 12. Tạo mảng 3 x 3 x 3 và tính tổng theo dòng, cột

```python
array_3d = np.random.randint(1, 10, size=(3, 3, 3))
print("Mảng 3x3x3:\n", array_3d)

# Tính tổng theo mỗi dòng (axis=2) và mỗi cột (axis=1) của từng ma trận con
sum_rows = np.sum(array_3d, axis=2)
sum_cols = np.sum(array_3d, axis=1)

print("\nTổng theo mỗi dòng (mỗi lớp):\n", sum_rows)
print("\nTổng theo mỗi cột (mỗi lớp):\n", sum_cols)
```

### 14. Cộng vector y vào từng dòng của ma trận A (Broadcasting)

```python
A = np.random.randint(1, 10, size=(4, 3))
y = np.array([10, 20, 30])

# NumPy tự động thực hiện broadcasting khi cộng ma trận 4x3 với vector 1x3
result_A = A + y

print("Ma trận A (4x3):\n", A)
print("Vector y (3 phần tử):", y)
print("Kết quả sau khi cộng y vào từng dòng:\n", result_A)
```

### 15. Cộng, trừ, nhân, chia 2 Pandas Series

```python
import pandas as pd

s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 10])

print("Cộng:\n", s1 + s2)
print("Trừ:\n", s1 - s2)
print("Nhân:\n", s1 * s2)
print("Chia:\n", s1 / s2)
```

### 16. Chuyển đổi cột đầu tiên của DataFrame thành Series

```python
df = pd.DataFrame({
    'Column1': [100, 200, 300],
    'Column2': ['A', 'B', 'C']
})

# Lưu ý: Thuộc tính .ix đã bị loại bỏ ở các phiên bản Pandas mới.
# Ta sử dụng .iloc hoặc truy cập trực tiếp bằng tên cột.
first_col_series = df.iloc[:, 0]

print("DataFrame:\n", df)
print("\nCột đầu tiên dưới dạng Series:\n", first_col_series)
print("Kiểu dữ liệu:", type(first_col_series))
```
